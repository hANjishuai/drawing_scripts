```python
import pandas as pd
import scanpy as sc
```

```python
#设置环境变量
sc.settings.verbosity = 3  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor="white")
#读取h5ad文件
results_file = "../../output/02.cluster/"  # the file that will store the analysis results
adata = sc.read_h5ad("../../output/01.QC/3.B1.QC_failtosct_V5.h5ad")
#基因去重
adata.var_names_make_unique()  # this is unnecessary if using `var_names='gene_ids'` in `sc.read_10x_mtx`
#相当于findhvg
sc.pl.highest_expr_genes(adata, n_top=20)
#数据清洗1
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
```

```python
#数据清洗2
adata.var["mt"] = adata.var_names.str.startswith("MT-")
sc.pp.calculate_qc_metrics(
    adata, qc_vars=["mt"], percent_top=None, log1p=False, inplace=True
)

sc.pl.violin(
    adata,
    ["n_genes_by_counts", "total_counts", "pct_counts_mt"],
    jitter=0.001,
    multi_panel=True,groupby="condition"
)

sc.pl.scatter(adata, x="total_counts", y="pct_counts_mt")
sc.pl.scatter(adata, x="total_counts", y="n_genes_by_counts")

adata = adata[adata.obs.n_genes_by_counts < 5000, :]
adata = adata[adata.obs.pct_counts_mt < 4, :].copy()
```

```python
#相当于seurat流程的NFS
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.highly_variable_genes(adata)

adata.raw = adata

adata = adata[:, adata.var.highly_variable]
sc.pp.regress_out(adata, ["total_counts", "pct_counts_mt"])
sc.pp.scale(adata, max_value=10)
#PCA
sc.tl.pca(adata, svd_solver="arpack")
sc.pl.pca(adata, color="MS4A1")
sc.pl.pca_variance_ratio(adata, log=True,n_pcs=50)

adata.write(results_file+"HC_DLE_SLE_SKIN.h5ad")
adata
```

```python
#聚类分群
adata=sc.read_h5ad(results_file+"HC_DLE_SLE_SKIN.h5ad")
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)
sc.pl.umap(adata, color=["MS4A1", "CD79A", "JCHAIN"])
sc.tl.leiden(
    adata,
    resolution=0.2,
    random_state=0,
    flavor="igraph",
    n_iterations=2,
    directed=False,
)

sc.pl.umap(adata, color=["leiden", "MS4A1", "JCHAIN","MZB1"],ncols=1)

adata.write(results_file+"HC_DLE_SLE_SKIN.h5ad")
adata
```
